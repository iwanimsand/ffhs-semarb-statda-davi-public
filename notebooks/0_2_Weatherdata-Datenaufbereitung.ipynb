````python
def missing_values(data):
    values = {
        'Total': data.isna().sum().values,
        'Prozent': data.isna().sum().values / len(data) * 100,
    }
    return pd.DataFrame(data=values, index=data.columns.values)
````

````python
missing_values(weatherdata)
````

Darum werden die Werte noch nach Jahr gruppiert und summiert.

````python
weatherdata.loc[:, ['dt_utc', 'rain_1h', 'snow_1h']].groupby([weatherdata['dt_utc'].dt.year]).sum()
````

# Tägliche Zusammenfassung

Es sind nur Wetterdaten ab 2013 nötig, da es erst ab 2013 Daten zu CitiBike gibt. Daher reichen die Daten ab 2013 für die tägliche Zusammenfassung.

````python
weatherdata_2013 = weatherdata[weatherdata['dt_utc'] > '2013']
````

````python
weatherdata_idx = weatherdata_2013.set_index(pd.DatetimeIndex(data=weatherdata_2013['dt_utc'].dt.date, name='Date'))
````

````python
weatherdata_daily = (
    weatherdata_idx.groupby(by='Date')
    [['temp', 'humidity', 'wind_speed', 'rain_1h', 'snow_1h']]
    .agg({
         'temp': ['mean', 'std', 'median', 'min', 'max'],
         'humidity': ['mean', 'std', 'median', 'min', 'max'],
         'wind_speed': ['mean', 'std', 'median', 'min', 'max'],
         'rain_1h': [ 'sum', 'min', 'max'],
         'snow_1h': [ 'sum', 'min', 'max']
     })
)
````

````python
# flatten and rename columns
weatherdata_daily.columns = weatherdata_daily.columns.to_flat_index().map(lambda tup: ' '.join(tup))
````

````python
weatherdata_daily
````

````python
weatherdata_daily.to_parquet(os.path.join(path, 'openweathermap-weatherdata-daily.parquet'), index=True)
````

````python
weatherdata_daily.to_csv(os.path.join(path, 'openweathermap-weatherdata-daily.csv'), index=True)
````
